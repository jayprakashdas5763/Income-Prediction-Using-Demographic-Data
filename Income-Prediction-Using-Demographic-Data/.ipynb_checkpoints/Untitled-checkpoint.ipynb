```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance, PartialDependenceDisplay
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.feature_selection import mutual_info_classif
from scipy.stats import pointbiserialr
from pathlib import Path

sns.set(style="whitegrid", context="talk", palette="deep")
```

```python

# 1)
# Paths
root = Path("..").resolve() /"capstone_project"
raw_path = root / "data" / "raw"
processed_path = root / "data" / "processed"
processed_path.mkdir(parents=True, exist_ok=True)


print(f"[Info] Root: {root}")
print(f"[Info] Raw path exists? {raw_path.exists()}")
print(f"[Info] Processed path: {processed_path}")

# 2) Column names from adult.names
columns = [
   "age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
   "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
   "hours-per-week", "native-country", "income"
]

# 3) Load file
in_file = raw_path / "adult_data.csv"  # adjust as needed, e.g., 'adult.data'
if not in_file.exists():
   raise FileNotFoundError(f"Input file not found: {in_file}")

df = pd.read_csv(
   in_file,
   names=columns,
   sep=",",
   skipinitialspace=True,
   na_values=["?"],
   low_memory=False
)

# Trim whitespace in object columns
for col in df.select_dtypes(include="object").columns:
   df[col] = df[col].str.strip()


# Optional: normalize target if it has trailing dots (common in adult.test)
if "income" in df.columns and df["income"].dtype == "object":
   df["income"] = df["income"].str.replace(r"\.$", "", regex=True)

# Ensure numeric types
numeric_cols = ["age", "fnlwgt", "education-num", "capital-gain", "capital-loss", "hours-per-week"]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")


# 5) Quick sanity checks
print(f"[Info] Loaded rows: {df.shape[0]}, columns: {df.shape[1]}")
print("[Info] DataFrame head:")
print(df.head())

print("\n[Info] Missing values per column:")
print(df.isna().sum())

print("\n[Info] Target value counts (including NaN):")
print(df["income"].value_counts(dropna=False))


# Save as CSV
out_file = processed_path / "adult.csv"
df.to_csv(out_file, index=False)
print(f"[Info] Saved cleaned CSV to: {out_file}")

# Remove rows with Missing Values

df_before = df.shape[0]
df = df.dropna()
print(f"Dropped {df_before - df.shape[0]} rows with missing values. Remaining: {df.shape[0]}")

# Basic Schema & Target Prep
import html

# -- Adjust these names according to your Dataset Description --

TARGET_COL = "income"
HOURS_COL = "hours-per-week"
EDU_YEARS_COL = "education-num"
AGE_COL = "age"

# 1) Clean up the target: unescape HTML, strip spaces
df[TARGET_COL] = (
   df[TARGET_COL]
   .astype(str)
   .apply(html.unescape)       # '&gt;' -> '>'
   .str.strip()
   .str.replace(r"\.$", "", regex=True)  # remove trailing '.' if present

)

# 2) Map to binary without downcasting warning
mapping = {'>50K': 1, '<=50K': 0}
df[TARGET_COL] = df[TARGET_COL].map(mapping)


# 3) Check if anything remained unmapped (NaN)
invalid_mask = df[TARGET_COL].isna()
if invalid_mask.any():
   print("Unmapped target labels encountered (showing up to 10 examples):")
   # Show original raw values for debugging; if you have a backup column, use it.
   print(df.loc[invalid_mask, TARGET_COL].head(10))
   print("Counts of unmapped labels:")
   print(df.loc[invalid_mask, TARGET_COL].value_counts(dropna=False))

# 4) Drop rows with NaN in target before casting (recommended for modeling)
before = len(df)
df = df.dropna(subset=[TARGET_COL]).copy()
print(f"Dropped {before - len(df)} rows with NaN in target '{TARGET_COL}'.")

# 4) Explicit dtype (prevents future downcasting warnings)
df[TARGET_COL] = df[TARGET_COL].astype('int8')  # or 'int64' if you prefer
print("Target distribution:", df[TARGET_COL].value_counts())

# Identify numeric vs categorical features
num_cols = df.select_dtypes(include=['int64', 'float64', 'int32', 'float32', 'int8']).columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Remove target from feature lists
num_cols = [c for c in num_cols if c != TARGET_COL]
cat_cols = [c for c in cat_cols if c != TARGET_COL]
```

```python

# 2) 
# =========================
# Remove Outliers
#    - IQR rule for numeric columns (except target)
#    - Sanity checks on HOURS & AGE
# =========================
def iqr_filter(data, col, k=1.5):
   Q1, Q3 = data[col].quantile([0.25, 0.75])
   IQR = Q3 - Q1
   lower = Q1 - k*IQR
   upper = Q3 + k*IQR
   return data[(data[col] >= lower) & (data[col] <= upper)]

df_out = df.copy()

# Apply IQR filter for selected continuous variables
for col in [AGE_COL, EDU_YEARS_COL, HOURS_COL]:
   if col in df_out.columns and np.issubdtype(df_out[col].dtype, np.number):
       before = df_out.shape[0]
       df_out = iqr_filter(df_out, col, k=1.5)
       print(f"{col}: removed {before - df_out.shape[0]} outliers (IQR rule).")

# Additional sanity filters (optional/customize thresholds):
# Weekly hours: assume 1–100 reasonable
if HOURS_COL in df_out.columns:
   before = df_out.shape[0]
   df_out = df_out[(df_out[HOURS_COL] >= 1) & (df_out[HOURS_COL] <= 100)]
   print(f"{HOURS_COL} sanity filter: removed {before - df_out.shape[0]} rows outside 1–100 hours.")

# Age sanity: assume working-age range
if AGE_COL in df_out.columns:
   before = df_out.shape[0]
   df_out = df_out[(df_out[AGE_COL] >= 16) & (df_out[AGE_COL] <= 90)]
   print(f"{AGE_COL} sanity filter: removed {before - df_out.shape[0]} rows outside 16–90.")

print("Remaining rows after outlier removal:", df_out.shape[0])

# Save cleaned dataset

```

```python

# =========================
# Modeling Setup
#    - OneHot for categorical
#    - Standardize numerics
#    - Baseline classifier
# =========================
X = df_out.drop(columns=[TARGET_COL])
y = df_out[TARGET_COL]

numeric_transformer = Pipeline(steps=[
   ("scaler", StandardScaler())
])

categorical_transformer = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

preprocessor = ColumnTransformer(
   transformers=[
       ("num", numeric_transformer, [c for c in num_cols if c in X.columns]),
       ("cat", categorical_transformer, [c for c in cat_cols if c in X.columns])
   ],
   remainder="drop"
)

# Try a robust classifier
clf = Pipeline(steps=[
   ("preprocess", preprocessor),
   ("model", RandomForestClassifier(n_estimators=300, random_state=42, class_weight="balanced"))
])

X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.25, random_state=42, stratify=y
)

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
y_prob = clf.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python

#  4)
# =========================
# Permutation importance with Pipeline (original features)
# =========================
from sklearn.inspection import permutation_importance, PartialDependenceDisplay

# 1) Use original input column names
feature_names_orig = X_test.columns.to_list()

# 2) Permutation importance over the pipeline (shuffles original features)
r = permutation_importance(
   clf, X_test, y_test,
   n_repeats=10, random_state=42, scoring="roc_auc"
)

# 3) Build a matched-length DataFrame (no transformed names here)
importances_orig = pd.DataFrame({
   "feature": feature_names_orig,
   "importance_mean": r.importances_mean,
   "importance_std": r.importances_std
}).sort_values("importance_mean", ascending=False)

display(importances_orig.head(15))

# 4) Highlight weekly working hours
hours_imp_orig = importances_orig.loc[importances_orig["feature"] == HOURS_COL]
print("\nWeekly working hours importance (original feature space):\n", hours_imp_orig)


numeric_cols = ["age", "fnlwgt", "education-num", "capital-gain", "capital-loss", "hours-per-week"]
X_test = X_test.copy()
for col in numeric_cols:
   if col in X_test.columns:
       X_test[col] = pd.to_numeric(X_test[col], errors="coerce").astype(float)

# 5) Partial dependence for HOURS_COL (Pipeline supports original names)
fig, ax = plt.subplots(figsize=(7, 5))
try:
   PartialDependenceDisplay.from_estimator(
       clf, X_test, [HOURS_COL], kind="average", ax=ax
   )
   plt.title("Partial Dependence: Weekly Working Hours vs Earning Probability")
   plt.tight_layout()
   # Optional save
   from pathlib import Path
   VISUALS = Path("visuals")
   VISUALS.mkdir(parents=True, exist_ok=True)
   plt.savefig(VISUALS / "pdp_hours_per_week.png", dpi=150)
   plt.show()
except Exception as e:
   print("PDP could not be plotted for HOURS_COL:", e)

# 6) Save to reports
from pathlib import Path
REPORTS = Path("reports")
REPORTS.mkdir(parents=True, exist_ok=True)
importances_orig.to_csv(REPORTS / "permutation_importance_original.csv", index=False)
print("Saved:", REPORTS / "permutation_importance_original.csv")

```

```python

# =========================
# B) Permutation importance in transformed (OHE) space
# =========================
preproc = clf.named_steps["preprocess"]
model = clf.named_steps["model"]

# 1) Transform X_test into the numeric/OHE matrix
X_test_trans = preproc.transform(X_test)

# 2) Get transformed feature names
try:
   feature_names_trans = preproc.get_feature_names_out()
except Exception:
   # Fallback: reconstruct names manually
   num_features = preproc.transformers_[0][2]  # columns sent to 'num'
   cat_features = preproc.transformers_[1][2]  # columns sent to 'cat'
   ohe = preproc.named_transformers_["cat"]
   cat_ohe_names = ohe.get_feature_names_out(cat_features)
   feature_names_trans = np.concatenate(
       [np.array(num_features, dtype=object), cat_ohe_names]
   )

# Sanity check
assert X_test_trans.shape[1] == len(feature_names_trans), (
   f"Transformed shape {X_test_trans.shape[1]} != names {len(feature_names_trans)}"
)

# 3) Permutation importance on the model with transformed inputs
r2 = permutation_importance(
   model, X_test_trans, y_test,
   n_repeats=10, random_state=42, scoring="roc_auc"
)

importances_trans = pd.DataFrame({
   "feature": feature_names_trans,
   "importance_mean": r2.importances_mean,
   "importance_std": r2.importances_std
}).sort_values("importance_mean", ascending=False)

display(importances_trans.head(20))

# 4) Pull hours-per-week importance from transformed feature list (numeric block first)
hours_idx = None
try:
   num_features = preproc.transformers_[0][2]
   if HOURS_COL in num_features:
       hours_idx = num_features.index(HOURS_COL)  # position within numeric block
       hours_imp_trans = importances_trans.iloc[[hours_idx]]
       print("\nWeekly working hours importance (transformed space, numeric index-based):\n", hours_imp_trans)
   else:
       print(f"Note: {HOURS_COL} not in numeric features list: {num_features}")
except Exception as e:
   print("Could not locate HOURS_COL index in transformed space:", e)

# 5) Save
importances_trans.to_csv(REPORTS / "permutation_importance_transformed.csv", index=False)


# -----------------------------------------
# 1) Point-biserial correlation (numeric features only)
# -----------------------------------------
feature_names_num = numeric_cols  # numeric feature list used in preprocessing

# df_out should contain numeric features and the target for correlation
# We'll build it from X (original columns) and y
df_out = X.copy()
df_out[TARGET_COL] = y  # binary 0/1

pbs_results = []
for col in feature_names_num:
   try:
       # point-biserial requires numeric feature and binary target
       r, p = pointbiserialr(df_out[col].astype(float), df_out[TARGET_COL].astype(int))
       pbs_results.append((col, r, p))
   except Exception as e:
       # If any column has all-NaN or constant values, correlation fails; record NaN
       pbs_results.append((col, np.nan, np.nan))

pbs_df = (
   pd.DataFrame(pbs_results, columns=["feature", "point_biserial_r", "p_value"])
     .sort_values("point_biserial_r", ascending=False)
     .reset_index(drop=True)
)

print("[Info] Point-biserial correlation (numeric features):")
display(pbs_df)

# -----------------------------------------
# 2) Mutual information (after preprocessing)
# -----------------------------------------
# Transform X via the fitted preprocessor inside your pipeline
X_trans = clf.named_steps["preprocess"].transform(X)

# Get transformed feature names from the ColumnTransformer
# This works for StandardScaler (numeric) + OneHotEncoder (categorical)
try:
   feature_names_trans = clf.named_steps["preprocess"].get_feature_names_out()
except Exception:
   # Fallback: construct names manually if get_feature_names_out is unavailable
   # Numeric names: "num__<col>", Categorical names: "cat__<col>_<category>"
   feature_names_trans = []
   # Numeric transformer name and columns should match your ColumnTransformer
   feature_names_trans += [f"num__{col}" for col in numeric_cols]
   # For categorical, OneHotEncoder can provide categories per column after fit
   ohe = clf.named_steps["preprocess"].named_transformers_["cat"]
   cat_categories = ohe.categories_
   for col_name, cats in zip(categorical_cols, cat_categories):
       feature_names_trans += [f"cat__{col_name}_{cat}" for cat in cats]

# Compute MI
mi = mutual_info_classif(X_trans, y, random_state=42, discrete_features=False)
mi_df = (
   pd.DataFrame({"feature": feature_names_trans, "mutual_info": mi}).sort_values("mutual_info", ascending=False).reset_index(drop=True)
)

print("\n[Info] Mutual Information (top 20 transformed features):")
display(mi_df.head(20))

# -----------------------------------------
# 3) Save reports
# -----------------------------------------
pbs_path = REPORTS / "point_biserial_correlations.csv"
mi_path  = REPORTS / "mutual_information_features.csv"
pbs_df.to_csv(pbs_path, index=False)
mi_df.to_csv(mi_path, index=False)
print(f"[Info] Saved:\n - {pbs_path}\n - {mi_path}")

```

```python

# =========================
# 5) Education Years vs Earning Potential
#    - Visuals + simple logistic with single predictor
# =========================
fig, ax = plt.subplots(1,2, figsize=(12,5))
sns.boxplot(data=df_out, x=TARGET_COL, y=EDU_YEARS_COL, ax=ax[0])
ax[0].set_title("Education years vs Income class")

sns.histplot(data=df_out, x=EDU_YEARS_COL, hue=TARGET_COL, stat="density", common_norm=False, ax=ax[1])
ax[1].set_title("Distribution by Income")
plt.tight_layout()
plt.show()

# Simple model with single feature to estimate odds ratio
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

X_edu = df_out[[EDU_YEARS_COL]]
y_bin = df_out[TARGET_COL]
edu_clf = make_pipeline(StandardScaler(), LogisticRegression(random_state=42))
edu_clf.fit(X_edu, y_bin)
coef = edu_clf.named_steps["logisticregression"].coef_[0][0]
odds_ratio = np.exp(coef)

```

```python

# =========================
# 6) Age vs Earning Potential
#    - Visuals + simple logistic with single predictor
# =========================
fig, ax = plt.subplots(1,2, figsize=(12,5))
sns.boxplot(data=df_out, x=TARGET_COL, y=AGE_COL, ax=ax[0])
ax[0].set_title("Age vs Income class")

sns.histplot(data=df_out, x=AGE_COL, hue=TARGET_COL, stat="density", common_norm=False, ax=ax[1])
ax[1].set_title("Age Distribution by Income")
plt.tight_layout()
plt.show()

X_age = df_out[[AGE_COL]]
age_clf = make_pipeline(StandardScaler(), LogisticRegression(random_state=42))
age_clf.fit(X_age, y_bin)
coef_age = age_clf.named_steps["logisticregression"].coef_[0][0]
odds_ratio_age = np.exp(coef_age)
print(f"Estimated odds ratio per 1 std increase in {AGE_COL}: {odds_ratio_age:.3f}")
```

```python

# =========================
# Wrap-up: Save Selected Outputs
# =========================
# Save the clean dataset and a quick summary
summary = {
   "rows_after_cleaning": df_out.shape[0],
   "roc_auc_random_forest": float(roc_auc_score(y_test, y_prob)),
   "hours_feature_in_dataset": HOURS_COL in df_out.columns,
   "education_years_feature_in_dataset": EDU_YEARS_COL in df_out.columns,
   "age_feature_in_dataset": AGE_COL in df_out.columns
}
pd.Series(summary).to_csv(REPORTS /"summary_metrics.csv")
print("Saved: cleaned_dataset.csv, point_biserial_correlations.csv, mutual_information_features.csv, summary_metrics.csv")
```
